```python
# Setup Jupyter environment and load in useful packages 
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import G
```

```python
from TidalPy.planets import build_planet, scale_from_planet

# First lets build a reference object
earth = build_planet('Earth')
_ = earth.paint()
```

```python
import pprint
pprint.pprint(earth.config)
```

```python
# Make a few exoplanets that are different scales of earth
exoplanet_radii_scale = [0.9, 1.0, 1.1, 1.2, 1.3]
exoplanet_build_info = [(scale, f'R={scale:0.1f}Earth') for scale in exoplanet_radii_scale]
exoplanets = {new_name: scale_from_planet(earth, radius_scale=scale, new_name=new_name) for scale, new_name in exoplanet_build_info}
```

```python
# Print some information about the new planets
for exoplanet_name, exoplanet in exoplanets.items():
    print(f'Radius Fraction = {exoplanet.radius / earth.radius:0.2f}')
    print(f'Mass Fraction = {exoplanet.mass / earth.mass:0.2f}')
    _ = exoplanet.paint()
    print()
```
