```python
import numpy as np
import matplotlib.pyplot as plt

density = np.linspace(1., 20000., 300)
period = np.linspace(0.01, 100., 400)
frequency = np.pi * 2 / (86400. * period)

den_mat, freq_mat = np.meshgrid(density, frequency)

term = freq_mat**2 * density
term_inv = 1. / term
```

```python
fig, ax = plt.subplots()

cdata = ax.contourf(density, period, np.log10(term), 50)
ax.set(xlabel='Density [kg m-3]', ylabel='Period [Days]')
cbar = fig.colorbar(cdata)
cbar.set_label('$(\\rho \\omega^2)$ Log10')

plt.show()
```
