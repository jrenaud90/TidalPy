## Build Input Files
The FE element model expects PREM-like input files that have even radial spacing. 
We specifically want to look at cases where TidalPy may differ due to additional physics.
Also want to look at the Shida Love number descrepancy we have found when comparing to LoadDef. 

```python
import numpy as np
import pandas as pd
from TidalPy.utilities.conversions import rads2days, days2rads
from TidalPy.rheology import Maxwell, Elastic, Newton
from loaddef import radial_solver_loaddef

min_shells_per_layer = 5

cases = dict()

# Homogeneous body with rocky density
case_name = 'homogeneous'
radius = 2000.0e3
shells = 50
cases[case_name] = dict()
cases[case_name]['frequency'] = days2rads(1.0)
cases[case_name]['radius_array'] = np.linspace(0.0, radius, shells, dtype=np.float64)
cases[case_name]['density_array'] = 4500.0 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_modulus'] = 25.0e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['bulk_modulus'] = 125.0e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_viscosity'] = 1.0e20 * np.ones(shells, dtype=np.float64)
cases[case_name]['bulk_viscosity'] = 1.0e40 * np.ones(shells, dtype=np.float64)
cases[case_name]['layer_upper_radius'] = (radius,)
cases[case_name]['layer_type'] = ('solid',)
cases[case_name]['shear_rheology_model'] = (Maxwell(),)
cases[case_name]['bulk_rheology_model'] = (Elastic(),)

# Large homogeneous body with rocky density
case_name = 'large_homogeneous'
radius = 6000.0e3
shells = 50
cases[case_name] = dict()
cases[case_name]['frequency'] = days2rads(1.0)
cases[case_name]['radius_array'] = np.linspace(0.0, radius, shells, dtype=np.float64)
cases[case_name]['density_array'] = 4500.0 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_modulus'] = 25.0e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['bulk_modulus'] = 125.0e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_viscosity'] = 1.0e20 * np.ones(shells, dtype=np.float64)
cases[case_name]['bulk_viscosity'] = 1.0e40 * np.ones(shells, dtype=np.float64)
cases[case_name]['layer_upper_radius'] = (radius,)
cases[case_name]['layer_type'] = ('solid',)
cases[case_name]['shear_rheology_model'] = (Maxwell(),)
cases[case_name]['bulk_rheology_model'] = (Elastic(),)

# Large homogeneous body with density near compressibility instability
case_name = 'unstable_homogeneous'
radius = 6000.0e3
shells = 50
cases[case_name] = dict()
cases[case_name]['frequency'] = days2rads(1.0)
cases[case_name]['radius_array'] = np.linspace(0.0, radius, shells, dtype=np.float64)
cases[case_name]['density_array'] = 4750.0 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_modulus'] = 25.0e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['bulk_modulus'] = 50.0e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_viscosity'] = 1.0e20 * np.ones(shells, dtype=np.float64)
cases[case_name]['bulk_viscosity'] = 1.0e40 * np.ones(shells, dtype=np.float64)
cases[case_name]['layer_upper_radius'] = (radius,)
cases[case_name]['layer_type'] = ('solid',)
cases[case_name]['shear_rheology_model'] = (Maxwell(),)
cases[case_name]['bulk_rheology_model'] = (Elastic(),)

# Liquid Core with higher density core
case_name = 'liquid_core'
radius = 2000.0e3
shells = 50
cases[case_name] = dict()
cases[case_name]['frequency'] = days2rads(1.0)
cases[case_name]['radius_array'] = np.linspace(0.0, radius, shells, dtype=np.float64)
cases[case_name]['density_array'] = 4500.0 * np.ones(shells, dtype=np.float64)
cases[case_name]['density_array'][cases[case_name]['radius_array'] < radius * 0.3] = 6000.0
cases[case_name]['shear_modulus'] = 25.0e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_modulus'][cases[case_name]['radius_array'] < radius * 0.3] = 0.0
cases[case_name]['bulk_modulus'] = 125.0e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['bulk_modulus'][cases[case_name]['radius_array'] < radius * 0.3] = 100.0e9
cases[case_name]['shear_viscosity'] = 1.0e20 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_viscosity'][cases[case_name]['radius_array'] < radius * 0.3] = 1000.0
cases[case_name]['bulk_viscosity'] = 1.0e40 * np.ones(shells, dtype=np.float64)
cases[case_name]['layer_upper_radius'] = (radius * 0.3, radius)
cases[case_name]['layer_type'] = ('liquid', 'solid',)
cases[case_name]['shear_rheology_model'] = (Newton(), Maxwell())
cases[case_name]['bulk_rheology_model'] = (Elastic(), Elastic())

# Icy world with ocean
case_name = 'icy_ocean'
radius = 2000.0e3
shells = 50
cases[case_name] = dict()
cases[case_name]['frequency'] = days2rads(1.0)
cases[case_name]['radius_array'] = np.linspace(0.0, radius, shells, dtype=np.float64)
cases[case_name]['density_array'] = 916.0 * np.ones(shells, dtype=np.float64)
cases[case_name]['density_array'][cases[case_name]['radius_array'] < radius * 0.85] = 1000.0
cases[case_name]['density_array'][cases[case_name]['radius_array'] < radius * 0.75] = 4500.0
cases[case_name]['shear_modulus'] = 3.3e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_modulus'][cases[case_name]['radius_array'] < radius * 0.85] = 0.0
cases[case_name]['shear_modulus'][cases[case_name]['radius_array'] < radius * 0.75] = 25.0e9
cases[case_name]['bulk_modulus'] = 8.0e9 * np.ones(shells, dtype=np.float64)
cases[case_name]['bulk_modulus'][cases[case_name]['radius_array'] < radius * 0.85] = 2.2e9
cases[case_name]['bulk_modulus'][cases[case_name]['radius_array'] < radius * 0.75] = 125.0e9
cases[case_name]['shear_viscosity'] = 1.0e16 * np.ones(shells, dtype=np.float64)
cases[case_name]['shear_viscosity'][cases[case_name]['radius_array'] < radius * 0.85] = 1000.0
cases[case_name]['shear_viscosity'][cases[case_name]['radius_array'] < radius * 0.75] = 1.0e20
cases[case_name]['bulk_viscosity'] = 1.0e40 * np.ones(shells, dtype=np.float64)
cases[case_name]['layer_upper_radius'] = (radius * 0.75, radius * 0.85, radius)
cases[case_name]['layer_type'] = ('solid', 'liquid', 'solid',)
cases[case_name]['shear_rheology_model'] = (Maxwell(), Newton(), Maxwell())
cases[case_name]['bulk_rheology_model'] = (Elastic(), Elastic(), Elastic())

# For all cases, save as a PREM like file. 
for case_name, case_data in cases.items():
    radius_array = case_data['radius_array']
    vs = np.sqrt(case_data['shear_modulus'] / case_data['density_array'])
    vp = np.sqrt((case_data['bulk_modulus'] + (4.0/3.0) * case_data['shear_modulus']) / case_data['density_array'])
    
    df = pd.DataFrame({
        'radius': case_data['radius_array'],
        'density': case_data['density_array'],
        'pwave_v': vp,
        'swave_v': vs
    })
    
    # Save to CSV
    # index=False prevents pandas from adding a 0,1,2... row index
    df.to_csv(f'fem_{case_name}.csv', index=False, float_format='%.10e')
```

## TidalPy results_by_subcase_by_subcase_by_subcase
Love number results using TidalPy

```python
from TidalPy.RadialSolver import build_rs_input_from_data, radial_solver

rs_kwargs = dict(
    degree_l = 2,
    solve_for = ('tidal', 'loading'),
    starting_radius = 0.0,
    start_radius_tolerance = 1.0e-5,
    nondimensionalize = True,
    # Shooting method parameters
    use_kamata = True,
    integration_method = 'DOP853',
    integration_rtol = 1.0e-5,
    integration_atol = 1.0e-8,
    scale_rtols_bylayer_type = False,
    max_num_steps = 500_000,
    expected_size = 1000,
    max_ram_MB = 500,
    max_step = 0,
    # Propagation matrix method parameters
    use_prop_matrix = False,
    core_model = 0,
    # Equation of State solver parameters
    eos_method_bylayer = None,
    surface_pressure = 0.0,
    eos_integration_method = 'DOP853',
    eos_rtol = 1.0e-3,
    eos_atol = 1.0e-5,
    eos_pressure_tol = 1.0e-3,
    eos_max_iters = 40,
    # Error and log reporting
    verbose = False,
    warnings = True,
    raise_on_fail = False,
    perform_checks = True,
    log_info = False
)

# Sub cases explore if compressibility and dynamic tides are considered
subcases = dict(
    incomp__static = (False, False),
    comp____static = (True, False),
    comp___dynamic = (True, True),
    incomp_dynamic = (False, True)
)

results_by_case = dict()
loaddef_results_by_case = dict()
for case_name, case_data in cases.items():
    print(f"Working on case: {case_name}.")
    num_layers = len(case_data['layer_upper_radius'])

    results_by_subcase = dict()
    loaddef_results_by_subcase = dict()
    for subcase_name, (compressible, dynamic) in subcases.items():

        # TidalPy calculations
        incomp_by_layer = tuple([not compressible for i in range(num_layers)])
        static_by_layer = tuple([not dynamic for i in range(num_layers)])
        rs_input = build_rs_input_from_data(
            case_data['frequency'],
            case_data['radius_array'],
            case_data['density_array'],
            case_data['bulk_modulus'],
            case_data['shear_modulus'],
            case_data['bulk_viscosity'],
            case_data['shear_viscosity'],
            case_data['layer_upper_radius'],
            case_data['layer_type'],
            static_by_layer,
            incomp_by_layer,
            case_data['shear_rheology_model'],
            case_data['bulk_rheology_model'],
            perform_checks=True,
            warnings=False
        )
        results_by_subcase[subcase_name] = radial_solver(*rs_input, **rs_kwargs)
        if not results_by_subcase[subcase_name].success:
            print(f"\t{subcase_name} was not successful: {results_by_subcase[subcase_name].message}.")

        complex_bulk_modulus_array = results_by_subcase[subcase_name].bulk_modulus_array
        complex_shear_modulus_array = results_by_subcase[subcase_name].shear_modulus_array

        # LoadDef calculations
        try:
            ((k_ld, h_ld, l_ld), (k_load_ld, h_load_ld, l_load_ld)) = radial_solver_loaddef(
                radius_array,
                case_data['density_array'],
                complex_bulk_modulus_array,
                complex_shear_modulus_array,
                case_data['frequency'],
                degree_l = rs_kwargs['degree_l'],
                # Loaddef defaults
                inf_tol=1.0e-5,
                rel_tol=1.0e-13,
                abs_tol=1.0e-13,
                backend='dop853',
                nstps=3000,
                kx=1,
                num_soln=100,
                nmaxfull=None,
                eval_radii=[],
                )
            loaddef_results_by_subcase[subcase_name] = ((k_ld, h_ld, l_ld), (k_load_ld, h_load_ld, l_load_ld))
        except:
            # Loaddef failed to find a solution; store nans
            loaddef_results_by_subcase[subcase_name] = (([np.nan], [np.nan], [np.nan]), ([np.nan], [np.nan], [np.nan]))
    
    results_by_case[case_name] = results_by_subcase
    loaddef_results_by_case[case_name] = loaddef_results_by_subcase
    print("\tDone with calculations.\n")
    
    # Tidal
    print(f"\tTidal k_{rs_kwargs['degree_l']}")
    for subcase_name, rs_result in results_by_subcase.items():
        if rs_result.success:
            print(f"\t\t{subcase_name}: {np.real(rs_result.k[0]):0.4e} {np.imag(rs_result.k[0]):0.4e}.")
        else:
            print(f"\t\t{subcase_name}: Not Successful (May not be supported).")
    for subcase_name, ld_result in loaddef_results_by_subcase.items():
        print(f"\t\tLoadDef {subcase_name}: {ld_result[0][0][0]:0.4e}")

    print(f"\tTidal h_{rs_kwargs['degree_l']}")
    for subcase_name, rs_result in results_by_subcase.items():
        if rs_result.success:
            print(f"\t\t{subcase_name}: {np.real(rs_result.h[0]):0.4e} {np.imag(rs_result.h[0]):0.4e}.")
        else:
            print(f"\t\t{subcase_name}: Not Successful (May not be supported).")
    for subcase_name, ld_result in loaddef_results_by_subcase.items():
        print(f"\t\tLoadDef {subcase_name}: {ld_result[0][1][0]:0.4e}")

    print(f"\tTidal l_{rs_kwargs['degree_l']}")
    for subcase_name, rs_result in results_by_subcase.items():
        if rs_result.success:
            print(f"\t\t{subcase_name}: {np.real(rs_result.l[0]):0.4e} {np.imag(rs_result.l[0]):0.4e}.")
        else:
            print(f"\t\t{subcase_name}: Not Successful (May not be supported).")
    for subcase_name, ld_result in loaddef_results_by_subcase.items():
        print(f"\t\tLoadDef {subcase_name}: {ld_result[0][2][0]:0.4e}")

    # Loading
    print(f"\tLoading k_{rs_kwargs['degree_l']}")
    for subcase_name, rs_result in results_by_subcase.items():
        if rs_result.success:
            print(f"\t\t{subcase_name}: {np.real(rs_result.k[1]):0.4e} {np.imag(rs_result.k[1]):0.4e}.")
        else:
            print(f"\t\t{subcase_name}: Not Successful (May not be supported).")
    for subcase_name, ld_result in loaddef_results_by_subcase.items():
        print(f"\t\tLoadDef {subcase_name}: {ld_result[1][0][0]:0.4e}")

    print(f"\tLoading h_{rs_kwargs['degree_l']}")
    for subcase_name, rs_result in results_by_subcase.items():
        if rs_result.success:
            print(f"\t\t{subcase_name}: {np.real(rs_result.h[1]):0.4e} {np.imag(rs_result.h[1]):0.4e}.")
        else:
            print(f"\t\t{subcase_name}: Not Successful (May not be supported).")
    for subcase_name, ld_result in loaddef_results_by_subcase.items():
        print(f"\t\tLoadDef {subcase_name}: {ld_result[1][1][0]:0.4e}")

    print(f"\tLoading l_{rs_kwargs['degree_l']}")
    for subcase_name, rs_result in results_by_subcase.items():
        if rs_result.success:
            print(f"\t\t{subcase_name}: {np.real(rs_result.l[1]):0.4e} {np.imag(rs_result.l[1]):0.4e}.")
        else:
            print(f"\t\t{subcase_name}: Not Successful (May not be supported).")
    for subcase_name, ld_result in loaddef_results_by_subcase.items():
        print(f"\t\tLoadDef {subcase_name}: {ld_result[1][2][0]:0.4e}")
    print("\n\n\n")

# Save TidalPy data to file
header = ['case_name', 'sub_case_name']
data = list()
for case_i, (case_name, case_results) in enumerate(results_by_case.items()):
    for subcase_i, (subcase_name, subcase_results) in enumerate(case_results.items()):
        bad_row = False
        row_data = list()
        for love_type_i, love_type in enumerate(['tidal', 'loading']):
            for letter in ('k', 'h', 'l'):
                for complex_name, complex_func in (('real', np.real), ('imag', np.imag)):
                    if case_i == 0 and subcase_i == 0:
                        header.append(f'{love_type}_{letter}_l{rs_kwargs['degree_l']}_{complex_name}')
                    if np.isnan(np.real(getattr(subcase_results, 'k')[0])):
                        bad_row = True
                    else:
                        bad_row = False
                    row_data.append(f"{complex_func(getattr(subcase_results, letter)[love_type_i]):0.4e}")
        if not bad_row:
            row_data = [case_name, subcase_name] + row_data
            data.append(row_data)

with open("tidalpy_results.dat", 'w') as file:
    file.write(", ".join(header) + '\n')
    for row in data:
        file.write(", ".join(row) + '\n')

# Save LoadDef data to file
header = ['case_name', 'sub_case_name']
data = list()
for case_i, (case_name, case_results) in enumerate(loaddef_results_by_case.items()):
    for subcase_i, (subcase_name, subcase_results) in enumerate(case_results.items()):
        bad_row = False
        row_data = list()
        for love_type_i, love_type in enumerate(['tidal', 'loading']):
            for l_i, letter in enumerate(('k', 'h', 'l')):
                for complex_name, complex_func in (('real', np.real), ('imag', np.imag)):
                    if case_i == 0 and subcase_i == 0:
                        header.append(f'{love_type}_{letter}_l{rs_kwargs['degree_l']}_{complex_name}')
                    if np.isnan(subcase_results[0][0][0]):
                        bad_row = True
                    else:
                        bad_row = False
                    if complex_name == 'imag':
                        # LoadDef does not solve imaginary portion
                        row_data.append(f"{np.nan}")
                    else:
                        row_data.append(f"{subcase_results[love_type_i][l_i][0]:0.4e}")
        if not bad_row:
            row_data = [case_name, subcase_name] + row_data
            data.append(row_data)

with open("loaddef_results.dat", 'w') as file:
    file.write(", ".join(header) + '\n')
    for row in data:
        file.write(", ".join(row) + '\n')

                

        
```
