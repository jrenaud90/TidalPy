# Eccentricity and Obliquity Function Performance

## Obliquity (Inclination) Functions

```python
from TidalPy.Tides_x.obliquity import obliquity_func

obliquity = 0.7

for l in (2, 3, 10):
    print(f"\nDegree l = {l}.")
    for trunc in ('gen', 2, 4, 'off'):
        print(f"\tTrunc = {trunc}.")
        %timeit obliquity_func(obliquity, l, trunc)
# Results as of TidalPy 0.7.1x
# Degree l = 2.
# 	Trunc = gen.
# 2.36 μs ± 30.5 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
# 	Trunc = 2.
# 2.21 μs ± 22.6 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
# 	Trunc = 4.
# 2.31 μs ± 20.7 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
# 	Trunc = off.
# 1.94 μs ± 9.98 ns per loop (mean ± std. dev. of 7 runs, 1,000,000 loops each)
# Degree l = 3.
# 	Trunc = gen.
# 2.6 μs ± 55.7 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
# 	Trunc = 2.
# 2.39 μs ± 25.7 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
# 	Trunc = 4.
# 2.56 μs ± 48.3 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
# 	Trunc = off.
# 1.94 μs ± 12.2 ns per loop (mean ± std. dev. of 7 runs, 1,000,000 loops each)
# ...
# Degree l = 10.
# 	Trunc = gen.
# 5.22 μs ± 67.4 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
# 	Trunc = 2.
# 3.94 μs ± 34.9 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
# 	Trunc = 4.
# 4.07 μs ± 37.2 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
# 	Trunc = off.
# 2.92 μs ± 36.6 ns per loop (mean ± std. dev. of 7 runs, 100,000 loops each)
```
